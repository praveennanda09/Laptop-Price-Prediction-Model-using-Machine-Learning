```python
def visualize_box_plots(df, column_list):
    # Set the size of the plot
    plt.figure(figsize=(15, len(column_list) *2))

    # Loop through each column in the list
    for i, column in enumerate(column_list):
        plt.subplot(len(column_list), 1, i + 1)  # Create subplots for each column
        sns.boxplot(x=df[column], color='skyblue')  # Create a boxplot for the current column
        plt.title(f'Box Plot of {column}', fontsize=14)
        plt.xlabel(column, fontsize=12)
        plt.tight_layout()  # Ensure that subplots are neatly arranged

    # Show the plots
    plt.show()


# Example usage:
column_list = ['Inches', 'CPU_Frequency (GHz)', 'RAM (GB)','Weight (kg)']
visualize_box_plots(df, column_list)
```

```python
Q1 = df['Inches'].quantile(0.25)  # 25th percentile
Q3 = df['Inches'].quantile(0.75)  # 75th percentile
IQR = Q3 - Q1  # Interquartile range

# 2. Defining the Lower and Upper Limits for Outliers
lower_limit = Q1 - (1.5 * IQR)  # Lower limit for outliers
upper_limit = Q3 + (1.5 * IQR)  # Upper limit for outliers

# 3. Capping the Outliers
df2 = df.copy()  # Create a copy of the original DataFrame

# Cap values below the lower limit to the lower limit
df2['Inches'] = np.where(df2['Inches'] < lower_limit, lower_limit, df2['Inches'])

# Cap values above the upper limit to the upper limit
df2['Inches'] = np.where(df2['Inches'] > upper_limit, upper_limit, df2['Inches'])

# 4. Print the new shape of the DataFrame
print("Shape after handling outliers: ", df2.shape)

# 5. Plotting the Boxplot to visualize the 'Inches' column after outlier handling
fig, axes = plt.subplots(1, 1, figsize=(6, 4))  # Adjust figure size
sns.boxplot(df2['Inches'], orient='v', ax=axes)
axes.set_title("Boxplot of 'Inches' after outlier capping")
plt.show()

df['Inches'] =df2['Inches']
```

```python
Q1 = df['CPU_Frequency (GHz)'].quantile(0.25)  # 25th percentile
Q3 = df['CPU_Frequency (GHz)'].quantile(0.75)  # 75th percentile
IQR = Q3 - Q1  # Interquartile range

# 2. Defining the Lower and Upper Limits for Outliers
lower_limit = Q1 - (1.5 * IQR)  # Lower limit for outliers
upper_limit = Q3 + (1.5 * IQR)  # Upper limit for outliers

# 3. Capping the Outliers
df2 = df.copy()  # Create a copy of the original DataFrame

# Cap values below the lower limit to the lower limit
df2['CPU_Frequency (GHz)'] = np.where(df2['CPU_Frequency (GHz)'] < lower_limit, lower_limit, df2['CPU_Frequency (GHz)'])

# Cap values above the upper limit to the upper limit
df2['CPU_Frequency (GHz)'] = np.where(df2['CPU_Frequency (GHz)'] > upper_limit, upper_limit, df2['CPU_Frequency (GHz)'])

# 4. Print the new shape of the DataFrame
print("Shape after handling outliers: ", df2.shape)

# 5. Plotting the Boxplot to visualize the 'CPU_Frequency (GHz)' column after outlier handling
fig, axes = plt.subplots(1, 1, figsize=(6, 4))  # Adjust figure size
sns.boxplot(df2['CPU_Frequency (GHz)'], orient='v', ax=axes)
axes.set_title("Boxplot of 'CPU_Frequency (GHz)' after outlier capping")
plt.show()

df['CPU_Frequency (GHz)'] = df2['CPU_Frequency (GHz)']
```

## **Encoding**

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

df['Company'] = encoder.fit_transform(df['Company'])
df['Product'] = encoder.fit_transform(df['Product'])
df['TypeName'] = encoder.fit_transform(df['TypeName'])
df['ScreenResolution'] = encoder.fit_transform(df['ScreenResolution'])
df['CPU_Company'] = encoder.fit_transform(df['CPU_Company'])
df['CPU_Model'] = encoder.fit_transform(df['CPU_Model'])
df['Storage'] = encoder.fit_transform(df['Storage'])
df['GPU_Company'] = encoder.fit_transform(df['GPU_Company'])
df['Graphic_Card'] = encoder.fit_transform(df['Graphic_Card'])
df['OS'] = encoder.fit_transform(df['OS'])
```

```python
x = df.drop(['Price', 'Inches', 'Weight (kg)'], axis=1)
y = df.iloc[:, -1]


```

```python
x
```

## **Scaling**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()

y_df = y.to_frame()
x = scaler.fit_transform(x)
# Fit and transform y
y = scaler.fit_transform(y_df)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, shuffle=True, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
model_1 = LinearRegression()
model_1.fit(x_train,y_train)
```

```python
y_pred = model_1.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

y_test_binary = np.round(y_test).astype(int)
y_pred_binary = np.round(y_pred).astype(int)

accuracy = accuracy_score(y_test_binary, y_pred_binary)
print('Accuracy:', accuracy)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print('R-squared:', r2)
print('Mean Absolute Error:', mae)
print('Mean Squared Error:', mse)
print('Root Mean Squared Error:', rmse)
```

```python
# Predicting the values
y_pred = model_1.predict(x_train)

# Reshaping y_train and y_pred if needed
y_train = y_train.ravel()  # or .flatten() if it's 2D
y_pred = y_pred.ravel()    # same for y_pred

# Plotting the actual vs predicted values
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_train, y=y_pred, label='Predicted vs Actual')
sns.lineplot(x=y_train, y=y_train, color='red', label='Best Fit Line')

plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Linear Regression: Actual vs Predicted')
plt.legend()
plt.show()
```

```python
# Predicting the values
y_pred = model_1.predict(x_train)

# Plotting the actual vs predicted values and the best-fit line
plt.figure(figsize=(10, 6))

# Scatter plot for actual values
sns.scatterplot(x=x_train[:, 0], y=y_train.ravel(), label='Actual Values', color='blue')

# Line plot for predicted values (best-fitting line)
sns.lineplot(x=x_train[:, 0], y=y_pred.ravel(), label='Best Fit Line', color='red')

plt.xlabel('Features (x_train)')
plt.ylabel('Target (y_train and y_pred)')
plt.title('Linear Regression: Actual vs Predicted with Best Fit Line')
plt.legend()
plt.show()
```
